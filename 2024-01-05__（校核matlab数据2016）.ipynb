```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
import os
```

### 主要为我做的matlab数据以及我师兄的数据的校核

```python
os.listdir( '/mnt/e/_03_data_air')
```

```python
path_file1 = '/mnt/e/_03_data_air/gz_all_2019.xlsx' # (自定义)对比文件1 ##########################（old）
path_file2 = '/mnt/e/_03_data_air/station_2019.xlsx' # (自定义)对比文件2 ##############################（new）
sheets = ['2362A', '2363A', '2364A', '2365A', '2366A']
```

```python
### 读取文件
old = pd.read_excel(path_file1,sheet_name='2363A')
new = pd.read_excel(path_file2,sheet_name='2363A')

### 设置画布和子图
fig, axes = plt.subplots(nrows=1, figsize=(16, 4))

### y轴数据点
yy1 = np.nan_to_num(new['PM2.5']) - np.nan_to_num(old['PM2.5'])
yy2 = np.nan_to_num(new['PM10']) - np.nan_to_num(old['PM10'])
yy3 = np.nan_to_num(new['SO2']) - np.nan_to_num(old['SO2'])
yy4 = np.nan_to_num(new['NO2']) - np.nan_to_num(old['NO2'])
yy5 = np.nan_to_num(new['O3']) - np.nan_to_num(old['O3'])
yy6 = np.nan_to_num(new['O3_8h']) - np.nan_to_num(old['O3_8h'])
yy7 = np.nan_to_num(new['CO']) - np.nan_to_num(old['CO'])

### x轴数据点
x = np.arange(0, 8784)

### 绘制差异线，使用不同颜色
plot1 = axes.plot(x, yy1, linestyle='-', color='red', label='PM2.5 difference')
plot2 = axes.plot(x, yy2, linestyle='-', color='blue', label='PM10 difference')
plot3 = axes.plot(x, yy3, linestyle='-', color='green', label='SO2 difference')
plot4 = axes.plot(x, yy4, linestyle='-', color='orange', label='NO2 difference')
plot5 = axes.plot(x, yy5, linestyle='-', color='purple', label='O3 difference')
plot6 = axes.plot(x, yy6, linestyle='-', color='brown', label='O3_8h difference')
plot7 = axes.plot(x, yy7, linestyle='-', color='pink', label='CO2 difference')

### 设置标题
axes.set_title('Difference Lines of 2363A')

### x轴和y轴网格线
axes.yaxis.grid(visible=True, color='black', linestyle='--')
axes.xaxis.grid(visible=True, color='gray', linestyle=':')

### x轴刻度和刻度标签设置
axes.set_xticks(np.arange(0, 8800, step=250))
axes.set_xticklabels(np.arange(0, 8800, step=250),rotation = 45, horizontalalignment='right',fontsize=10)

### 设置图例
axes.legend(loc='upper left')
```

### 新版校核工具

```python
total = 0
# 初始化字典用于存储计数
count_dict = {}
for sheet in sheets:
    df1 = pd.read_excel(path_file1, sheet_name=sheet)
    df2 = pd.read_excel(path_file2, sheet_name=sheet)
    col_name = ['PM2.5', 'PM10', 'SO2', 'NO2', 'O3', 'O3_8h','CO']

    for col in col_name:
        for row in df1.index:
            a = df1[col][row]
            b = df2[col][row]
            c = df1[col][row]
            d = df2[col][row]

            if not (np.isnan(a) and np.isnan(b)) and a != b:
                print(f'在sheet {sheet}，对比文件1中 {c} 和对比文件2中 {d} 不同，在excel表行 {row + 2}，列 {col}。')
                total += 1

                # 根据站点和变量类型构建字典的键
                key = f'{sheet}_{col}'

                # 判断键是否存在于字典中，如果不存在则初始化为0
                count_dict[key] = count_dict.get(key, 0) + 1

# 输出每个站点和变量类型的不同数量
for key, value in count_dict.items():
    print(f'{key} 有 {value} 处不同')
# 输出总的不同数量
print(f'一共有 {total} 处不同')
```
